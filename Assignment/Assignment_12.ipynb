#### Ques 1. In what modes should the PdfFileReader() and PdfFileWriter() File objects will be opened?

Answer:

For PdfFileReader() file objects should be opened in rb -> read binary mode, whereas for PdfFileWriter() file objects should be opened in wb -> write binary mode.

#### Ques 2. From a PdfFileReader object, how do you get a Page object for page 5?

Answer:

PdfReader() class provides a method called getPage(page_no) to get a page object.

```python
import PyPDF2

pdf_file_path = 'D:/ineuron_notes/stats2.pdf'
page_number = 5

with open(pdf_file_path, 'rb') as file:
    pdf_reader = PyPDF2.PdfReader(file)
    if page_number < 1 or page_number > len(pdf_reader.pages):
        raise ValueError("Invalid page number.")
    page = pdf_reader.pages[page_number - 1]

print(page)
```

### Ques 3. What PdfFileReader variable stores the number of pages in the PDF document?

Answer: 

PdfReader() method of PdfFileReader class stores the no pages in a PDF document.

```python
import PyPDF2

pdf_file_path = 'D:/ineuron_notes/stats2.pdf'

with open(pdf_file_path, 'rb') as file:
    pdf_reader = PyPDF2.PdfReader(file)
    num_pages = len(pdf_reader.pages)

print(f'The PDF document has {num_pages} pages.')
```

#### Ques 4. If a PdfFileReader object’s PDF is encrypted with the password swordfish, what must you do before you can obtain Page objects from it?

Answer:

If a PdfFileReader object’s PDF is encrypted with the password swordfish and you're not aware of it. First read the Pdf using the PdfFileReader Class. #### Ques 5. What methods do you use to rotate a page?

Answer:

PyPDF2 Package provides 2 methods to rotate a page:

- rotateClockWise() -> For Clockwise rotation
- rotateCounterClockWise() -> For Counter Clockwise rotation

The PyPDF2 package only allows you to rotate a page in increments of 90 degrees. You will receive an AssertionError otherwise.

#### Ques 6. What is the difference between a Run object and a Paragraph object?

Answer:

The structure of a document is represented by three different data types in python-Docx. At the highest level, a Document object represents the entire document. The Document object contains a list of Paragraph objects for the paragraphs in the document. (A new paragraph begins whenever the user presses ENTER or RETURN while typing in a Word document.) Each of these Paragraph objects contains a list of one or more Run objects.

The text in a Word document is more than just a string. It has font, size, color, and other styling information associated with it. A style in Word is a collection of these attributes. A Run object is a contiguous run of text with the same style. A new Run object is needed whenever the text style changes.

#### Ques 10. How do you create a Document object for a new Word document?

```python
# Answer:

from docx import Document

# Create a new Document object
doc = Document()

# Add content to the document
doc.add_heading('My New Document', level=1)
doc.add_paragraph('This is the first paragraph.')
doc.add_paragraph('This is the second paragraph.')

# Save the document to a file
doc.save('my_new_document.docx')
```
